## Topic Modelling

### Preprocessing

```python
nike_reviews_df['reviewText'] = nike_reviews_df['reviewText'].map(lambda x: x.lower()).copy()
```

#### Stop Word Removal

```python
import nltk
nltk.download('stopwords')
```

```python
stop_words = stopwords.words('english')
```

```python
nike_reviews_df['reviewText'] = nike_reviews_df['reviewText'].apply(lambda tokenize_review: [w for w in tokenize_review if w not in stop_words])
nike_reviews_df
```

#### Phrase Modeling : Bigram Models

```python
# Build the bigram and trigram models
bigram = gensim.models.Phrases(nike_reviews_df['reviewText'], min_count=5, threshold=100) # higher threshold fewer phrases.
bigram_mod = gensim.models.phrases.Phraser(bigram)
nike_reviews_df['reviewText'] = nike_reviews_df['reviewText'].map(lambda tokenize_data: bigram_mod[tokenize_data])
```

```python
nike_reviews_df.head()
```

#### Lemmatization

```python
def lemmatization(tokenize_data, allowed_postags=['NOUN', 'ADJ', 'VERB', 'ADV']):
    doc = nlp(" ".join(tokenize_data)) 
    
    return [token.lemma_ for token in doc if token.pos_ in allowed_postags]
```

```python
# Initialize spacy 'en' model, keeping only tagger component (for efficiency)
nlp = spacy.load("en_core_web_sm", disable=['parser', 'ner'])
```

```python
nike_reviews_df['reviewText'] = nike_reviews_df['reviewText'].apply(lambda bigram_tokenize_review: lemmatization(bigram_tokenize_review))
```

#### Transformation: Corpus and Dictionary

```python
import gensim.corpora as corpora

# Create Dictionary
id2word = corpora.Dictionary(nike_reviews_df['reviewText'])

# Create Corpus
texts = nike_reviews_df['reviewText']

# Term Document Frequency
corpus = [id2word.doc2bow(text) for text in texts]

# View
print(corpus[:6])
```

### Modeling

```python
# Build LDA model
lda_model = gensim.models.LdaMulticore(corpus=corpus,
                                       id2word=id2word,
                                       num_topics=10, 
                                       random_state=100,
                                       chunksize=1000,
                                       passes=10,
                                       alpha=0.31,
                                       eta=0.7,
                                       per_word_topics=True)
```

### Evaluation & Hyperparameter Tuning (e.g Number of Topics(k), etc)

 Hyperparameters: 
- Number of Topics (`k`)
- Dirichlet hyperparameter `alpha`: Document-Topic Density
- Dirichlet hyperparameter `eta`: Word-Topic Density

```python
from gensim.models import CoherenceModel

# Compute Coherence Score
coherence_model_c_v = CoherenceModel(model=lda_model, texts=nike_reviews_df['reviewText'], dictionary=id2word, coherence='c_v')
coherence_model_u_mass = CoherenceModel(model=lda_model, corpus=corpus, dictionary=id2word, coherence='u_mass')
coherence_c_v = coherence_model_c_v.get_coherence()
print('Coherence Score: ', coherence_c_v)

coherence_u_mass = coherence_model_u_mass.get_coherence()
print('Coherence Score: ', coherence_u_mass)
```

```python
def compute_coherence_values(corpus, dictionary, k, alpha, eta):
    lda_model = gensim.models.LdaMulticore(corpus=corpus,
                                           id2word=dictionary,
                                           num_topics=k, 
                                           random_state=123321,
                                           chunksize=1000,
                                           passes=10,
                                           alpha=alpha,
                                           eta=eta)
    
    coherence_model = CoherenceModel(model=lda_model, texts=nike_reviews_df['reviewText'], 
                                     dictionary=id2word, coherence='c_v')    
    
    return coherence_model.get_coherence()
```

```python
import numpy as np
import tqdm

grid = {}
grid['Validation_Set'] = {}

# Topics range
min_topics = 8
max_topics = 43
step_size = 1
topics_range = range(min_topics, max_topics, step_size)

# Alpha parameters
alphas = list(np.arange(0.31, 1, 0.3))

# Eta parameters
etas = list(np.arange(0.31, 1, 0.3))

# Validation sets
num_of_docs = len(corpus)
valid_set = gensim.utils.ClippedCorpus(corpus, int(num_of_docs*0.75))

results = {'Number of topics': [],
           'Alpha': [],
           'Eta': [],
           'Coherence': []}

# Can take a long time to run
pbar = tqdm.tqdm(total=(len(etas)*len(alphas)*len(topics_range)))

# iterate through number of topics
for k in topics_range:
    # iterate through alpha values
    for alpha in alphas:
        # iterare through beta values
        for eta in etas:
            # get the coherence score for the given parameters
            coherence = compute_coherence_values(corpus=valid_set, 
                                                  dictionary=id2word, 
                                                  k=k, 
                                                  alpha=alpha, 
                                                  eta=eta)
            # Save the model results
            results['Number of topics'].append(k)
            results['Alpha'].append(alpha)
            results['Eta'].append(eta)
            results['Coherence'].append(coherence)
            
            pbar.update(1)

pbar.close()

results = pd.DataFrame(results)            
results.to_csv('lda_tuning_results.csv', index=False)
```

## Marketing & Product Insights

### Demographic segmentation

Demographic segmentation is a crucial aspect of marketing, as different age groups and genders have unique needs and preferences. For instance, men may be interested in shoes that are versatile and can be used for multiple purposes, such as gym shoes (topic 0) and shoes that are suitable for any occasion (topic 3). In contrast, children may require shoes that are both comfortable and visually appealing (topic 4), while women may prioritize shoes that are both worthy and fashionable (topic 13). Additionally, gentlemen may seek shoes that convey strength and confidence (topic 5). Understanding the distinctive needs of these demographic segments can aid businesses in tailoring their marketing strategies to better cater to their target audiences.

### What's not in the data

It seems that one missing topic is the opinions of long-time Nike shoe users. There may be many reviews from customers who have purchased Nike shoes multiple times because of their positive experiences with the brand. These customers may have insights into how Nike shoes have evolved over time, and may comment on whether the shoes have maintained their quality and durability. Additionally, long-time users may have a deeper appreciation for the design and functionality of Nike shoes, and may be able to speak to how different models compare to each other.
